# Persist extents from a Shapefile into an ExtentStoreNC object

```python
from awrams.utils import geo, gis, extents
from awrams.utils import mapping_types as mt
from awrams.utils.io import data_mapping as dm
import numpy as np
```

```python
nces = gis.ExtentStoreNC('calibration_extents_test_5k.nc','w')
```

```python
e = extents.get_default_extent()
```

```python
estore = gis.ExtentStoreShapefile(gis.CATCHMENT_SHAPEFILE,'StationID',extents.get_default_extent())
```

```python
for k in estore.available:
    print(k)
    nces[k] = estore[k]
```

```python
emap = {}
for k in nces.available:
    emap[k] = nces[k]
```
